```python
import json
import pandas as pd
small_wine_df= pd.read_csv('wine-ratings-small.csv')
small_wine_df
```

```python
def cut_columns(df, column_list, index_name):
    '''This function extracts the specified columns in column_list and resets the index to the column index_name.'''
    
    df1=df.loc[:, column_list].copy()
    df1.set_index(index_name, inplace=True)
    return df1
    
```

```python
small_wine_df=cut_columns(small_wine_df, ['name', 'region', 'variety', 'rating', 'notes'], 'name')
small_wine_df
```

```python
small_wine_df=small_wine_df[small_wine_df.loc[:, 'rating']>=94].copy()
small_wine_df
```

```python
small_wine_dict=small_wine_df.to_dict('index')
small_wine_dict
```

```python
with open('best_wine_small.json', 'w') as f:
    json.dump(small_wine_dict, f)
```
